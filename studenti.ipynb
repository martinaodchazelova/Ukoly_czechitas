```python
# Frekvenční analýza pro kategorické sloupce
kategoricke_sloupce = ['Sex', 'Year', 'Faculty', 'Accommodation', 'Scholarship', 
                       'Study_Hours', 'Socialising_Frequency', 'Drinks_Per_Night', 
                       'Missed_Classes', 'Modules_Failed', 'In_Relationship', 
                       'Parents_Approve', 'Relationship_Parents']

# Zobrazení četností pro každou kategorickou proměnnou
for sloupec in kategoricke_sloupce:
    print(f"Frekvenční analýza pro {sloupec}:")
    print(data[sloupec].value_counts(dropna=False), "\n")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Vizualizace četností pro každou kategorickou proměnnou
for sloupec in kategoricke_sloupce:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=data, x=sloupec, order=data[sloupec].value_counts().index)
    plt.title(f'Četnosti kategorií: {sloupec}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
kategoricke_sloupce = ['Sex', 'Year', 'Faculty', 'Accommodation', 'Scholarship', 
                       'Study_Hours', 'Drinks_Per_Night', 'Missed_Classes', 
                       'Modules_Failed', 'In_Relationship', 'Parents_Approve', 
                       'Relationship_Parents']

# Vytvoření křížových tabulek
for kategoricka in kategoricke_sloupce:
    print(f"Křížová tabulka mezi Socialising_Frequency a {kategoricka}:")
    print(pd.crosstab(data['Socialising_Frequency'], data[kategoricka], margins=True), "\n")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Numerické sloupce, které budeme zkoumat
numeric_cols = ['Matric_GPA', 'Yearly_GPA']

# Boxploty pro Socialising_Frequency a numerické proměnné
for numeric_col in numeric_cols:
    plt.figure(figsize=(10, 5))
    sns.boxplot(data, x='Socialising_Frequency', y=numeric_col)
    plt.title(f'{numeric_col} podle Socialising_Frequency')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Boxplot pro zobrazení rozdílů v Yearly_GPA podle Socialising_Frequency a Sex
plt.figure(figsize=(12, 6))
sns.boxplot(data, x='Socialising_Frequency', y='Yearly_GPA', hue='Sex')
plt.title('Yearly_GPA podle Socialising_Frequency (rozlišení podle pohlaví)')
plt.xticks(rotation=45)
plt.legend(title='Sex')
plt.show()
```

```python

# Barplot pro 'Missed_Classes' podle 'Socialising_Frequency' a 'Sex'
plt.figure(figsize=(10, 6))
sns.countplot(data, x='Socialising_Frequency', hue='Missed_Classes', palette='muted')
plt.title('Distribuce Missed_Classes podle Socialising_Frequency')
plt.xticks(rotation=45)
plt.legend(title='Missed_Classes')
plt.show()
```
